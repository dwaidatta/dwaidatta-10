## Using Dataset from Kaggle

```python
# using pandas to create dataframe from dataset csv

import pandas as pd
df = pd.read_csv('world_population.csv')
print(df)
```

```python
# checking cols and datatypes
print(df.columns)
print(df.dtypes)
```

```python
# enhancing plot visuals

print(plt.style.available) # checking available options

# plt.style.use('dark_background') # using a style
```

```python
# sorting as per rank

df_sorted = df.sort_values('Rank')
print(df_sorted)
```

```python
# line plot: growth rate vs rank

plt.figure(figsize=(20,12)) # set the fig size in inches (width, height)
plt.plot(df_sorted['Rank'], df_sorted['Growth Rate'], color='green', marker='o', linestyle='-',label='Growth Rate')

# df_sorted['Rank'] ------ x axis
# df_sorted['Growth Rate'] ------ y axis
# color sets color of graph
# marker sets marker for data points (x, s, d, o)
# linestyle (--, -, :, -.)

plt.title('Population Growth Rate by Rank', fontsize=16) # sets title

# fontsize parameter

plt.xlabel('Rank', fontsize=12) # sets xlabel
plt.ylabel('Growth rate %', fontsize=12) # sets ylabel
plt.grid(True) # toggles grid for reability
plt.legend() # shows color index

plt.show() # displays graph
```

```python
# creating a bar plot

# top 50 countries by 2022 population
top_countries = df.nlargest(50, '2022 Population') # gets top 50 countries

plt.figure(figsize=(20,12))
plt.plot(top_countries['Country/Territory'], top_countries['2022 Population'], color='green', marker='o') # plt.plot() creates line graph
plt.bar(top_countries['Country/Territory'], top_countries['2022 Population'], color='teal', alpha=0.5) # plt.bar()) creates bar graph

# alpha sets transparency

plt.title("Top 5 countries by 2022 population", fontsize=24)
plt.xlabel('Country', fontsize=16)
plt.ylabel('2022 Population', fontsize=16)
plt.xticks(rotation=-90) # rotate x label labels for better readability

plt.show()
```

```python
# creating a histogram

plt.figure(figsize=(20,12))
plt.hist(df['Growth Rate'], bins=15, color='orange', edgecolor='black', alpha=0.7)

# df['Growth Rate']: data for histogram
# bins=15: no of bins(intervals) in the histogram
# alpha sets transparency
# edge color sets color of bin edges for clear separation

plt.title('Histogram')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.grid(True)

plt.show()
```

```python
# scatter plot
# area vs population density

plt.figure(figsize=(20,12))

plt.scatter(df['Area (km²)'], df['Density (per km²)'], c='red', alpha=0.6, s=50)

# s sets size of points
# c sets color of points

plt.axis([0, 20000, 0, 10000]) # sets range of axes
# plt.axis([xmin, xmax, ymin, ymax])

plt.title('Area vs Population density')
plt.xlabel('Area')
plt.ylabel('Population density')
plt.grid(True)

plt.show()
```

```python
# use autoscale alternatively

plt.figure(figsize=(20,12))

plt.scatter(df['Area (km²)'], df['Density (per km²)'], c='red', alpha=0.6, s=50)
plt.grid(True)

plt.autoscale(enable=True, axis='both', tight=True) # using plt.autoscale()

plt.show()
```

```python
# box plot
# box plot for growth rate

plt.figure(figsize=(20,12))

plt.boxplot(df['Growth Rate'], notch = True, vert=True)

plt.show()
```

## Multi Plot

```python
# plotting multiple cols with different styles

plt.figure(figsize=(20,12))

plt.plot(df['2022 Population'], label='2022 Population', color='blue', marker='o', alpha=0.5)
plt.plot(df['2015 Population'], label='2015 Population', color='green', linestyle='--', marker='x', alpha=0.5)

plt.title('Poputation over time', fontsize=20)
plt.xlabel('Index')
plt.ylabel('Population')

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(20,12))

# showing above graph as scatter

plt.scatter(df.index, df['2022 Population'], label='2022 Population', c='red', alpha=0.5, s=40)
plt.scatter(df.index, df['2015 Population'], label='2015 Population', c='green', alpha=0.5, s=40)

# plt.autoscale(enable=True,axis='both', tight=True)

plt.title('Poputation over time', fontsize=20)
plt.xlabel('Index')
plt.ylabel('Population')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# creating subplots
# create subplots for population and density

fig, axes = plt.subplots(1, 2, figsize=(20,12)) # 1 row, 2 col

# subplots creates a grid for subplots
# fig is the figure object
# axes is a list/array of subplot axes

# first subplot for population
axes[0].plot(df['2022 Population'], color='red')
axes[0].set_title('2022 Population')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Population')

# second subplot for density
axes[1].plot(df['Density (per km²)'], color='purple')
axes[1].set_title('Density')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Density')

plt.tight_layout() # prevent overlap [adjusts spacing between the plots]

plt.show()
```

```python
# customizing legends

plt.figure(figsize=(20,12))

plt.plot(df['2022 Population'], label='Population (2022)', color='blue', linestyle='-', marker='o', alpha=0.5)
plt.plot(df['Growth Rate'], label='Growth Rate', color='orange', linestyle='--', marker='x', alpha=0.5)

# customize the legend
plt.legend(loc='upper left', fontsize=10, title='Legend', shadow=True, fancybox=True)

plt.title('Population and Growth Rate')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

```python
# combining different plot types

plt.figure(figsize=(20,12))

plt.plot(df['2022 Population']*0.00000001, color='blue') # line plot

plt.bar(df.index, df['World Population Percentage'], label='World %', color='green', alpha=0.7) # bar plot

plt.title('Population vs World Percentage')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

plt.show()
```

```python
# twin axes
# doing the above graph using twin axes for proper scaling

fig, ax1 = plt.subplots(figsize=(20,12))

# first axis
ax1.plot(df.index, df['2022 Population'], color='blue', label='2022 Population', alpha=0.7) # line
ax1.set_ylabel('2022 Population', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(0)

# second axis
ax2 = ax1.twinx()
ax2.bar(df.index, df['World Population Percentage'], label='World %', color='green', alpha=0.5) # bar
ax2.set_ylabel('World Population Percentage', color='green')
ax2.tick_params(axis='y', labelcolor='green')

fig.suptitle('2022 Population and World Population Percentage by Index', fontsize=28)

plt.show()
```

```python
# logarithmic scale

plt.figure(figsize=(20,12))

plt.bar(df['Rank'], df['World Population Percentage'], color='lightpink')

plt.yscale('log')

plt.show()
```

```python
# showing the same with linear and log scale

fig, ax1 = plt.subplots(figsize=(20,12))

# linear axis
ax1.scatter(df['Rank'], df['World Population Percentage'], color='purple', alpha=0.5)
ax1.set_ylabel('Linear', color='purple')

# log axis
ax2 = ax1.twinx()
ax2.bar(df['Rank'], df['World Population Percentage'], color='pink', alpha=0.3)
ax2.set_ylabel('Log', color='pink')
ax2.set_yscale('log')

plt.show()
```

```python
# customizing ticks

top_countries = df.nlargest(50, '2022 Population') # gets top 50 countries

plt.figure(figsize=(20,12))
plt.plot(top_countries['Country/Territory'], top_countries['2022 Population'], color='green', marker='o') # plt.plot() creates line graph
plt.bar(top_countries['Country/Territory'], top_countries['2022 Population'], color='teal', alpha=0.5) # plt.bar()) creates bar graph

# alpha sets transparency

plt.title("Top 5 countries by 2022 population", fontsize=16)
plt.xlabel('Country')
plt.ylabel('2022 Population')

# ticks
plt.xticks(fontsize=16, rotation=-90)
plt.yticks(fontsize=16)

plt.grid(axis='y', linestyle='--', alpha=0.5, color='blue')

plt.show()
```

```python
# saving figures

# sample plot

plt.figure(figsize=(20,12))

plt.plot(df['2022 Population'], label='2022 Population', color='blue', marker='o', alpha=0.5)
plt.plot(df['2015 Population'], label='2015 Population', color='green', linestyle='--', marker='x', alpha=0.5)

plt.title('Poputation over time', fontsize=20)
plt.xlabel('Index')
plt.ylabel('Population')

plt.legend()
plt.grid(True)

# saving
plt.savefig("test_graph.png", dpi=300, bbox_inches='tight')

# dpi: resolution of saved image
# bbox_inches='tight': ensures no clipping of labels

plt.show()
```

# At A Glance

## **Plot Types**
| **Plot Type**               | **Function / Method**                | **Description**                                                   | **Syntax**                               | **Example**                                                                                      |
|------------------------------|--------------------------------------|-------------------------------------------------------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Line Plot**               | `plt.plot()`                        | Default plot type to visualize trends.                            | `plt.plot(x, y, label='Label')`         | `plt.plot(df['Rank'], df['2022 Population'], label='Population')`                              |
| **Bar Plot**                | `plt.bar()`                         | Vertical bar plots for categories.                                | `plt.bar(x, height)`                    | `plt.bar(df['Country/Territory'], df['2022 Population'])`                                        |
| **Horizontal Bar Plot**     | `plt.barh()`                        | Horizontal bar plots for categories.                              | `plt.barh(y, width)`                    | `plt.barh(df['Country/Territory'], df['2022 Population'])`                                       |
| **Scatter Plot**            | `plt.scatter()`                     | Points plotted to show relationships.                            | `plt.scatter(x, y, c='color')`          | `plt.scatter(df['Density (per km²)'], df['Growth Rate'], color='green')`                        |
| **Histogram**               | `plt.hist()`                        | Distribution of numerical data.                                   | `plt.hist(x, bins=10)`                  | `plt.hist(df['2022 Population'], bins=15, color='purple')`                                      |
| **Fill Between Lines**      | `plt.fill_between()`                | Highlight area between curves.                                    | `plt.fill_between(x, y1, y2)`           | `plt.fill_between(df['Rank'], y1, y2, color='blue', alpha=0.5)`                                 |

---

## **Customizing Axes**
| **Feature**                 | **Function / Method**                | **Description**                                                   | **Syntax**                               | **Example**                                                                                      |
|------------------------------|--------------------------------------|-------------------------------------------------------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Set X/Y Limits**          | `plt.xlim()` / `plt.ylim()`         | Adjust the range of x and y axes.                                 | `plt.xlim(left, right)`                 | `plt.xlim(0, 50)`                                                                                  |
| **Logarithmic Scale**        | `plt.yscale()`                      | Set logarithmic scale for y-axis.                                 | `plt.yscale('log')`                     | `plt.yscale('log')`                                                                              |
| **Customizing Ticks**        | `plt.xticks()` / `plt.yticks()`     | Format axis ticks.                                                | `plt.xticks(rotation=45)`               | `plt.xticks(fontsize=12, rotation=45)`                                                          |

---

## **Adding Information**
| **Feature**                 | **Function / Method**                | **Description**                                                   | **Syntax**                               | **Example**                                                                                      |
|------------------------------|--------------------------------------|-------------------------------------------------------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Labels**                  | `plt.xlabel()` / `plt.ylabel()`     | Add labels to x and y axes.                                       | `plt.xlabel('X-axis Label')`            | `plt.xlabel('Population Rank')`                                                                 |
| **Title**                   | `plt.title()`                       | Add a title to the plot.                                          | `plt.title('Title Here')`               | `plt.title('Population Trends')`                                                                |
| **Legend**                  | `plt.legend()`                      | Add a legend to the plot.                                         | `plt.legend(loc='best')`                | `plt.legend(loc='upper right')`                                                                 |
| **Annotations**             | `plt.annotate()`                    | Add annotations to a plot.                                        | `plt.annotate('Text', xy=(x, y))`       | `plt.annotate('Max', xy=(5, 1000), arrowprops=dict(arrowstyle='->'))`                          |

---

## **Layouts and Subplots**
| **Feature**                 | **Function / Method**                | **Description**                                                   | **Syntax**                               | **Example**                                                                                      |
|------------------------------|--------------------------------------|-------------------------------------------------------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Multiple Subplots**        | `plt.subplot()`                     | Create multiple subplots in one figure.                           | `plt.subplot(nrows, ncols, index)`      | `plt.subplot(2, 2, 1)`                                                                          |

---

## **Twin Axes**
| **Feature**                 | **Function / Method**                | **Description**                                                   | **Syntax**                               | **Example**                                                                                      |
|------------------------------|--------------------------------------|-------------------------------------------------------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Twin Axes**               | `ax1.twinx()`                       | Create a second y-axis sharing the same x-axis.                   | `ax2 = ax1.twinx()`                     | Combine growth rate and population in one graph.                                                 |

---

## **Saving Plots**
| **Feature**                 | **Function / Method**                | **Description**                                                   | **Syntax**                               | **Example**                                                                                      |
|------------------------------|--------------------------------------|-------------------------------------------------------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Save Figure**             | `plt.savefig()`                     | Save a figure to an external file.                                | `plt.savefig(filename, dpi)`            | `plt.savefig('output_plot.png', dpi=300, bbox_inches='tight')`                                  |
